1. Загрузите датасет с помощью функции np.genfromtxt()

```python
import numpy as np
import matplotlib.pyplot as plt

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = np.genfromtxt(url, delimiter=',', dtype='str', encoding='utf-8')
#names=['sepallength', 'sepalwidth', 'petallength', 'petalwidth', 'species']
iris [: 15]
```

2. Вычислите простейшие статистики: min, max, median, mean, 0.25-percentile,0.75-percentile для каждого из числовых атрибутов в датасете

```python
char = []
for i in range(4):
    char.append(np.genfromtxt(url, delimiter=',', dtype=None, encoding='utf-8', usecols=[i]))

for j in range(len(char)): 
    min, max = np.min(char[j]), np.max(char[j])
    med, mean = np.median(char[j]), round(np.mean(char[j]), 2) 
    perc25, perc75 = np.percentile(char[j],25), np.percentile(char[j],75)
    print(f'Мин. знач. характеристики {j} = ', min)
    print(f'Макс. знач. характеристики {j} = ', max)
    print(f'Медиан. знач. характеристики {j} = ', med)
    print(f'Сред. знач. характеристики {j} = ', mean)
    print(f'Перс. 0,25 характеристики {j} = ', perc25)
    print(f'Перс. 0,75 характеристики {j} = ', perc75)
    print ('\n')        
```

3. Постройте гистограмму для каждого из числовых атрибутов в датасете

```python
fig, ((graph1, graph2), (graph3, graph4)) = plt.subplots(nrows=2, ncols=2)

graph1.hist(char[0], color='orange', alpha=0.7)
graph1.set_title('sepallength (char0)')

graph2.hist(char[1], color='green', alpha=0.7)
graph2.set_title('sepalwidth (char1)')

graph3.hist(char[2], color='purple', alpha=0.7)
graph3.set_title('petallength (char2)')

graph4.hist(char[3], color='red', alpha=0.7)
graph4.set_title('petalwidth (char3)')

fig.tight_layout()
plt.show()
```

4. Сгруппируйте датасет по категории (последний столбец датасета) и повторитешаги 2 и 3 для групп.

```python
iris_unique= np.unique(iris[:,4])

for i in range(len(iris_unique)):
    iris_gr = iris[np.where(iris[:,4] == iris_unique[i])]
    for j in range(4):
        min_gr = round(np.min((iris_gr[:, j].astype('float')), axis = 0),2)
        max_gr = round(np.max((iris_gr[:, j].astype('float')), axis = 0),2)
        med_gr = round(np.median((iris_gr[:, j].astype('float')), axis = 0),2)
        mean_gr = round(np.mean((iris_gr[:, j].astype('float')), axis = 0),2)
        perc25_gr = round(np.percentile((iris_gr[:, j].astype('float')), 25, axis = 0),2)
        perc75_gr = round(np.percentile((iris_gr[:, j].astype('float')), 75, axis = 0),2)
        print('\033[94m'+'\033[3m' + 'Вид ириса:',iris_unique[i] + '\033[0m')
        print(f'Мин. знач. характеристики {j} = ', min_gr)
        print(f'Макс. знач. характеристики {j} = ', max_gr)
        print(f'Медиан. знач. характеристики {j} = ', med_gr)
        print(f'Средн. знач. характеристики {j} = ', mean_gr)
        print(f'Перс. 0,25 характеристики {j} = ', perc25_gr)
        print(f'Перс. 0,75 характеристики {j} = ', perc75_gr)
        plt.hist((iris_gr[:, j].astype('float')))
        plt.title(f'Гистограмма для характеристики {j}')
        plt.show()
        print ('\n') 
```
