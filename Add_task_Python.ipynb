1.Напишите программу, которая найдет все такие числа, которые делятся на 7, но не делятся на 5, от 2000 до 3200 (включая оба). Полученные числа должны быть напечатаны в последовательности, разделенной запятыми, в одной строке.

```python
a = range(2000, 3201)
result = []
for i in a:
  if i % 7 == 0 and i % 5 != 0: 
    result.append(i)
print("Полученные числа:", result)
print("Количество чисел:", len(result))   

```

2. Напишите программу, которая может вычислить факториалы заданных чисел. Результаты должны быть напечатаны в последовательности, разделенной запятыми, в одной строке.

```python
numbers = [1, 5, 4, 9, 10, 7]
factorial = []
for i in numbers:
  if i <= 0 or type(i) == float:
    print ('Введено некорректное число:',i)
  else:
    multiply = 1 
    for j in range(1, i+1):
      multiply = multiply*j
    factorial.append(multiply)
print("Факториалы заданных чисел:", factorial)
```

3.Напишите программу, которая принимает 2 числа X, Y в качестве входных данных с клавиатуры и генерирует двумерный массив. Значение элемента в i-й строке и j-м столбце массива должно быть i*j.  (нумерация с нуля)



```python
def martix_fun (x,y): 
    d1 = []
    for i in range (x):
      d = []
      for j in range (y):
        r = i*j
        d.append (r)
      d1.append(d)
    print (*d1, sep=',\n')

print('Укажите размерность матрицы')
print('Кол-во строк:')
x = int(input())
print('Кол-во столбцов:')
y = int(input())
martix_fun(x,y)
```

*   Как минимум 1 буква между [a-z]
*   По крайней мере, 1 число между [0–9]
*   По крайней мере, 1 буква между [A–Z]
*   По крайней мере, 1 символ из [$#@]

*   Максимальная длина пароля транзакции: 12




```python
import re
def password (w):
    if len(w) < 6:
      print ("Ошибка (пароль меньше 6 символов)")
    elif len(w) > 12:
      print ("Ошибка (пароль больше 12 символов)")
    elif not re.search ("[a-z]", w):
      print ("Ошибка (пароль не содержит строчные буквы a-z)")
    elif not re.search ("[0-9]", w):
      print ("Ошибка (пароль не содержит цифры 0-9)")
    elif not re.search ("[A-Z]", w):
      print ("Ошибка (пароль не содержит заглавные буквы A-Z)")
    elif not re.search ("[$#@]", w):
      print ("Ошибка (пароль не содержит символы $#@)")
    else:
      print ('Пароль подходит')

#print('Введите пароль:', 
w = str(input("Введите пароль: "))
print ()
print('Результат проверки пароля: ')
password (w) 
```
